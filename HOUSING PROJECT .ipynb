##  How to plot Histogram

```python
plt.figure(figsize=(7,7))
plt.hist(df["area"],bins=10,)
plt.title(["area frequency Distribution"])
plt.xlabel("area")
plt.ylabel("frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.hist(df["area"],bins=10,orientation='horizontal')
plt.title(["area frequency Distribution"])
plt.xlabel("area")
plt.ylabel("frequency")
plt.grid(True)
plt.show()
```

##  Histogram using pandas

```python
plt.figure(figsize=(7,7))
df['area'].hist(bins=10)
```

```python
w_mean=df['stories'].mean()
w_median=df['stories'].median()

plt.figure(figsize=(10,10))
df['stories'].hist(bins=10)
plt.title("stories fequency distribution")
plt.axvline(w_mean,color='purple')
plt.axvline(w_median,color='black')
plt.xlabel('stories')
plt.ylabel('Frequency')
plt.show()
```

##  Bar Chart

```python
plt.figure(figsize=(7,7))
df['bedrooms'].value_counts().plot(kind='bar')
plt.xlabel('bedrooms')
plt.ylabel('Count')
plt.title('Count of Houses by bedrooms')
plt.show()
```
